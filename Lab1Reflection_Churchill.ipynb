```
import networkx as nx
import matplotlib.pyplot as plt
CM=nx.DiGraph()
CM.add_nodes_from(['Agile','Class','Waterfall','Use Case','SuperClass','Methods','SubClass','Activity Diagram','Attributes'])

CM.add_edges_from([('Agile','Waterfall',)],label='used for project development')
CM.add_edges_from([('Class','Attributes',)],label='might have')
CM.add_edges_from([('Activity Diagram','Use Case',)],label='used for modeling')
CM.add_edges_from([('SuperClass','SubClass',)],label='can have more than one')
CM.add_edges_from([('Class','Methods',)],label='might have')

edgeLabels=nx.get_edge_attributes(CM, 'label')
plt.figure(1,figsize=(8,8))
pos=nx.circular_layout(CM)
nx.draw(CM, pos,arrows=True, node_size=7500, node_color='violet', 
        with_labels=True, edge_color='red')
nx.draw_networkx_edge_labels(CM, pos, edge_labels=edgeLabels,label_pos=0.5, font_size=11)
plt.show()

```
