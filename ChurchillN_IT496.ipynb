```python
from matplotlib import colors
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

#upload the ITSkills.csv to the Google Drive
df=pd.read_csv("/content/drive/MyDrive/ITSkills.csv")
print(df)

#set this column as index - to use as plot labels
df.index = df['skill']
#descending sort
df = df.sort_values(by='importance', ascending=False)

#comulative percentage column is created using cumsum function
df["comulativePercentage"] = df["importance"].cumsum()/df["importance"].sum()+100

#making sure only 2 decimal points are shown
df["comulativePercentage"]=df["comulativePercentage"].apply(lambda x: round(x,2))
print(df)

fig, ax = plt.subplots()
ax.bar(df.index, df["importance"], color="c0")

#show label diagonally
ax.set_xticklabels( df['skill'], rotation=45 )

#ax2 and ax1 will be shown at the same time with twinx
ax2 = ax.twinx()

ax2.plot(df.index, df["comulativePercentage"], color="C1", marker="D", ms=7)
ax2.yaxis.set_major_formatter(PercentFormatter())

ax.tick_params(axis="y", colors="C0")
ax2.tick_params(axis="y", colors="C1")

for index, v in df.iterrows():
  label = round(v['comulativePercentage'], 1)
  plt.annotate(label, xy=(v['skill'],v['comulativePercentage']))

plt.show()
```
